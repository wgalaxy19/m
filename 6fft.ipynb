```python
import numpy as np
import imageio
from scipy import fftpack
import matplotlib.pyplot as plt
image = imageio.imread('car.jpeg', mode='L')
# Perform 2D Fourier Transform
fft2d = fftpack.fft2(image)
# Shift zero frequency components to the center for better visualization
fft2d_shifted = fftpack.fftshift(fft2d)
# Compute the magnitude spectrum for 2D FFT
magnitude_spectrum_2d = np.log(np.abs(fft2d_shifted) + 1)
# Adding 1 to avoid
# Display the results
plt.figure(figsize=(12, 4))
# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
# 2D Fourier Transform
plt.subplot(1, 3, 2)
plt.imshow(magnitude_spectrum_2d, cmap='gray')
plt.title('2D Fourier Transform')
# Combined 1D Fourier Transforms along rows and columns
plt.subplot(1, 3, 3)
plt.plot(np.fft.fftfreq(image.shape[1]), np.log(np.abs(fftpack.fftshift(fftpack.fft(image,
axis=1))).mean(axis=0) + 1), label='Rows')
plt.plot(np.fft.fftfreq(image.shape[0]), np.log(np.abs(fftpack.fftshift(fftpack.fft(image,
axis=0))).mean(axis=1) + 1), label='Columns')
plt.title('Combined 1D Fourier Transforms')
plt.xlabel('Frequency')
plt.legend()
plt.show()
```
